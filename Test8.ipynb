```python
import os
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import cv2 as cv


# Define a matrix to determine the transformation
c = np.array([(50, 50), (50, 100), (150, 255), (150, 150)])

# Create the first linear transformation
t1 = np.linspace(0, c[0, 1], c[0, 0] + 1).astype('uint8')
print(len(t1))  # Print the length of t1

# Create the second linear transformation
t2 = np.linspace(c[0, 1] + 1, c[1, 1], c[1, 0] - c[0, 0]).astype('uint8')
print(len(t2))  # Print the length of t2

# Create the third linear transformation
t3 = np.linspace(c[1, 1] + 1, c[2, 1], c[2, 0] - c[1, 0]).astype('uint8')
print(len(t3))  # Print the length of t3

# Create the fourth linear transformation
t4 = np.linspace(c[2, 1] + 1, c[3, 1], c[3, 0] - c[2, 0]).astype('uint8')
print(len(t4))  # Print the length of t4

# Create the fifth linear transformation
t5 = np.linspace(c[3, 1] + 1, 255, 255 - c[3, 0]).astype('uint8')
print(len(t5))  # Print the length of t5

# Concatenate the five transformations to create the final 'transform' array
transform = np.concatenate((t1, t2, t3, t4), axis=0).astype('uint8')
transform = np.concatenate((transform, t5), axis=0).astype('uint8')
print(len(transform))  # Print the length of the final transform

# Visualize the transformation using a plot
fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input intensity')
ax.set_ylabel(r'Output intensity')
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_aspect('equal')
plt.savefig('transform.png')  # Save the plot as an image
plt.show()

# Mount Google Drive
drive.mount('/content/drive')

# Load the grayscale image using Matplotlib
img_path = '/content/drive/MyDrive/Photos/emma.jpg'
img_orig = plt.imread(img_path)

# Apply the defined transformation to the original image
image_transformed = np.take(transform, img_orig)

# Create a side-by-side comparison of original and transformed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display the original image in the first subplot
axes[0].imshow(img_orig, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

# Display the transformed image in the second subplot
axes[1].imshow(image_transformed, cmap='gray')
axes[1].set_title("Transformed Image")
axes[1].axis('off')

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load the image using OpenCV
img_path = '/content/drive/MyDrive/Photos/highlights_and_shadows.jpg'
img_bgr = cv.imread(img_path, cv.IMREAD_COLOR)

# Convert the BGR image to the L*a*b* color space
img_lab = cv.cvtColor(img_bgr, cv.COLOR_BGR2Lab)

# Extract the L channel
l_channel = img_lab[:, :, 0]

# Define the gamma value for gamma correction (specify your desired value)
gamma = 2.0

# Create a lookup table for gamma correction
table = np.array([(i / 255.0) ** (gamma) * 255.0 for i in np.arange(0, 256)]).astype('uint8')

# Apply the gamma correction to the L channel
l_channel_gamma = cv.LUT(l_channel, table)

# Replace the original L channel with the gamma-corrected L channel
img_lab[:, :, 0] = l_channel_gamma

# Convert the image back to the BGR color space
img_gamma_corrected = cv.cvtColor(img_lab, cv.COLOR_Lab2BGR)

# Display the original and gamma-corrected images
plt.figure(figsize=(12, 6))
plt.subplot(121), plt.imshow(cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)), plt.title("Original Image")
plt.subplot(122), plt.imshow(cv.cvtColor(img_gamma_corrected, cv.COLOR_BGR2RGB)), plt.title("Gamma Corrected Image")

# Define color channels to be plotted: blue, green, and red
color = ('b', 'g', 'r')

# Create two separate plots for histograms
plt.figure(figsize=(12, 6))

# Plot histograms for the original image
plt.subplot(121)
plt.title("Histograms for Original Image")
for i, c in enumerate(color):
    hist_orig = cv.calcHist([img_bgr], [i], None, [256], [0, 256])
    plt.plot(hist_orig, color=c, label=f'{c} channel')
plt.legend()

# Plot histograms for the gamma-corrected image
plt.subplot(122)
plt.title("Histograms for Gamma Corrected Image")
for i, c in enumerate(color):
    hist_gamma = cv.calcHist([img_gamma_corrected], [i], None, [256], [0, 256])
    plt.plot(hist_gamma, color=c, label=f'{c} channel')
plt.legend()

# State the gamma value used for correction
print(f"Gamma value used for correction: {gamma}")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/drive/MyDrive/Photos/spider.png'
img = cv2.imread(image_path)

# Convert the image to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Split the image into hue, saturation, and value planes
hue_plane = hsv_img[:, :, 0]
saturation_plane = hsv_img[:, :, 1]
value_plane = hsv_img[:, :, 2]

# Define the parameters
a = 0.7
sigma = 70.0

# Apply the intensity transformation to the saturation plane
transformed_saturation_plane = np.minimum(
    np.maximum(saturation_plane + a * 128 * np.exp(-((saturation_plane - 128) ** 2) / (2 * sigma ** 2)), 0), 255
).astype(np.uint8)

# Recombine the planes
vibrance_hsv = cv2.merge([hue_plane, transformed_saturation_plane, value_plane])

# Convert the HSV image back to BGR color space
vibrance_bgr = cv2.cvtColor(vibrance_hsv, cv2.COLOR_HSV2BGR)

# Display the images using Matplotlib
plt.figure(figsize=(12, 6))
plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title("Original Image")
plt.subplot(132), plt.imshow(cv2.cvtColor(vibrance_bgr, cv2.COLOR_BGR2RGB)), plt.title("Vibrance Enhanced Image")

# Create a visualization of the intensity transformation
x = np.arange(256)
intensity_transformation = np.minimum(
    np.maximum(x + a * 128 * np.exp(-((x - 128) ** 2) / (2 * sigma ** 2)), 0), 255
).astype(np.uint8)

plt.subplot(133), plt.plot(x, intensity_transformation, label='Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.legend()

plt.show()
```

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/drive/MyDrive/Photos/shells.tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Calculate the histogram and bins for the grayscale image
hist, bins = np.histogram(img.ravel(), 256, [0, 256])

# Calculate the cumulative distribution function (CDF) and normalize it
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# Plot the normalized CDF in blue and the histogram in red
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.plot(cdf_normalized, color='b')
plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.legend(('cdf', 'histogram'), loc='upper left')
plt.title('Histogram of the Original Image')

# Apply histogram equalization to the grayscale image
equ = cv2.equalizeHist(img)

# Calculate the histogram and bins for the equalized image
hist, bins = np.histogram(equ.ravel(), 256, [0, 256])

# Calculate the cumulative distribution function (CDF) and normalize it
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# Plot the normalized CDF in blue and the histogram in red for the equalized image
plt.subplot(132)
plt.plot(cdf_normalized, color='b')
plt.hist(equ.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.legend(('cdf', 'histogram'), loc='upper left')
plt.title('Histogram of the Equalized Image')

# Concatenate the original grayscale image and the equalized image side by side
res = np.hstack((img, equ))

# Display the concatenated images with no axis and grayscale colormap
plt.subplot(133)
plt.axis('off')
plt.imshow(res, cmap='gray')
plt.title('Original vs. Equalized Image')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/drive/MyDrive/Photos/jeniffer.jpg'
img = cv2.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Convert the image to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Split the image into hue, saturation, and value planes
hue_plane = hsv_img[:, :, 0]
saturation_plane = hsv_img[:, :, 1]
value_plane = hsv_img[:, :, 2]

# Display the planes in grayscale
plt.figure(figsize=(12, 4))
plt.subplot(131), plt.imshow(hue_plane, cmap='gray'), plt.title("Hue Plane")
plt.subplot(132), plt.imshow(saturation_plane, cmap='gray'), plt.title("Saturation Plane")
plt.subplot(133), plt.imshow(value_plane, cmap='gray'), plt.title("Value Plane")
plt.show()

# Threshold the value plane to obtain the foreground mask

threshold_value = 105
selected_plane = value_plane
_, mask = cv2.threshold(selected_plane, threshold_value, 255, cv2.THRESH_BINARY)

# Obtain the foreground using cv2.bitwise_and
foreground = cv2.bitwise_and(img, img, mask=mask)

# Compute the histogram of the masked region
hist_fg, bins_fg = np.histogram(selected_plane[mask > 0], 256, [0, 256])

# Calculate the cumulative sum of the histogram
cdf_fg = hist_fg.cumsum()

# Histogram equalization formula
cdf_min = cdf_fg.min()
num_pixels = len(selected_plane[mask > 0])
equalized_values = ((cdf_fg[selected_plane[mask > 0]] - cdf_min) * 255) / (num_pixels - 1)

# Replace the values in the selected plane with the equalized values
selected_plane[mask > 0] = equalized_values

# Invert the mask to get the background
background_mask = cv2.bitwise_not(mask)

# Create an equalized background plane
equalized_background = cv2.equalizeHist(selected_plane)

# Combine the histogram-equalized foreground and the equalized background
result_img = cv2.merge([hue_plane, saturation_plane, selected_plane])

# Convert the result back to BGR color space
result_img = cv2.cvtColor(result_img, cv2.COLOR_HSV2BGR)

# Display the hue, saturation, and value planes, the mask, the original image, and the result
# with the histogram-equalized foreground
plt.figure(figsize=(12, 8))
plt.subplot(231), plt.imshow(hue_plane, cmap='gray'), plt.title("Hue Plane")
plt.subplot(232), plt.imshow(saturation_plane, cmap='gray'), plt.title("Saturation Plane")
plt.subplot(233), plt.imshow(selected_plane, cmap='gray'), plt.title("Equalized Value Plane")
plt.subplot(234), plt.imshow(mask, cmap='gray'), plt.title("Foreground Mask")
plt.subplot(235), plt.imshow(img), plt.title("Original Image")
plt.subplot(236), plt.imshow(result_img), plt.title("Result with Histogram-Equalized Foreground")
plt.show()
```
